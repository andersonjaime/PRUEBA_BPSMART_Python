```python
import pandas as pd
import numpy as np
import datetime
```

```python
# Importar Dataframe mediante API
df_conv = pd.read_json('https://www.datos.gov.co/resource/33dq-ab5a.json')
```

```python
df_conv.head()
```

```python
df_conv2=df_conv.copy()
```

```python
# Convertir Dataset a Minúsculas
for col in df_conv2.columns:
  df_conv2[col] = df_conv2[col].astype(str).str.lower()
```

```python
df_conv2.head()
```

```python
# Quitar espacios
for col in df_conv2.columns:
  df_conv2[col] = df_conv2[col].str.replace(' ','')
```

```python
df_conv2.head()
```

```python
# Convertir Fechas a Timestamp (Se crean tres columnas nuevas para almacenar estos datos)

df_conv2["año"] = df_conv2["ano_convo"].dt.year

df_conv2["dia_semana"] = df_conv2["ano_convo"].dt.dayofweek

df_conv2["dia_semana_nombre"] = df_conv2["ano_convo"].dt.strftime("%A")

```

```python
df_conv2.head()
```

```python
# Dividir el nombre de la convocatoria en numero de convocatoria y año

df_conv2['nme_convocatoria']=df_conv2['nme_convocatoria'].str.replace('convocatoria','').str.replace('de','')
```

```python
df_conv2.head()
```

```python
#Dividir el Id del producto

df_conv2['id_producto_pd']=df_conv2['id_producto_pd'].str.replace('AP','').str.replace('ART','')
```

```python
df_conv2.head()
```

```python
# Eliminar Columnas Innecesarias (Se eliminan las nuevas columnas que se crearon anteriormente)

df_conv2 = df_conv2.drop(["año", "dia_semana", "dia_semana_nombre"], axis = 1)
```
